```python
import pyautogui
import time

# Tile positions relative to the screenshot region
tiles = [(50, 50), (150, 50), (230, 50), (300, 50)]
target_color = (0, 0, 0, 255)  # Target color (black tile with alpha channel)

# Region to capture: (x, y, width, height)
screenshot_region = (300, 400, 400, 100)

while True:
    # Take a screenshot of the defined region
    current_SS = pyautogui.screenshot(region=screenshot_region)
    
    # Check each tile's color
    for tile in tiles:
        pvalue = current_SS.getpixel(tile)
        
        if pvalue == target_color:
            # Calculate the absolute click position
            a, b = tile
            pyautogui.click(screenshot_region[0] + a, screenshot_region[1] + b)
            #break  # Break to avoid multiple clicks in one iteration
    
    # Optional delay to reduce CPU usage
    # time.sleep(0.01)
```
